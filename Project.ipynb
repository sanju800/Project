```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
myexcel=pd.read_csv('Desktop/Entri/myexcel - myexcel.csv.csv')
myexcel
```

1) Changing the column 'Height' having the incorrect data with any random numbers between 150 and 180

```python
myexcel['Height']=np.random.randint(150,180,size=len(myexcel))
```

```python
print("Modified 'Height' column")
print(myexcel['Height'])
```

2) How many are there in each team and the percentage splitting with respect to the total employees

```python
team_count=myexcel['Team'].value_counts()
percentage=(team_count/len(myexcel))*100
summary=pd.DataFrame({'Count':team_count,'Percentage':percentage})
summary
```

3) Segregate the employees w.r.t different positions

```python
c=myexcel['Position'].value_counts()
c
```

4) Find from which age group most of the employees belongs to 

```python
age_count=myexcel['Age'].value_counts()
age_count
```

5) Find out under which team and position, spending in terms of salary is high

```python
salary_spending=myexcel.groupby(['Team','Position'])['Salary'].sum().idxmax()
salary_spending
```

6) Find if there is any correlation between Age and Salary, represent it visually

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Salary',data=myexcel)
plt.title('Correlation between Age and Salary')
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```
